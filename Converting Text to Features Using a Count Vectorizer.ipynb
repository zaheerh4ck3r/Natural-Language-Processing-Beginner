```python
#Its used to show frequency
#A count vectorizer is similar to one-hot encoding, but instead of checking whether a
#particular word is present or not, it counts the words that are present in the document.
```

```python
from sklearn.feature_extraction.text import CountVectorizer
text = ["Its gonna be Legen wait for it Legen keep waiting Dary Legendary"]
vectorizer = CountVectorizer()
```

```python
vectorizer.fit(text)
```

```python
vector = vectorizer.transform(text)
```

```python
print(vectorizer.vocabulary_)
print(vector.toarray())
```
